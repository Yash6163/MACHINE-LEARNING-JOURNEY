# TRAIN DATA TEST DATA SPLIT


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42 )

# x mein se test size 25 percent nikla aur vo gya x_test and y_test mein

# matlab mera 0.25 ie 25 % hogya test data and 75 % hoga train data
```

# CONVERT IN STANDARDISATION
- DONE BY USING Z SCORE MEAN=0 STD. DEVN=1 
- AS WEIGHT IS IN (KG) AND HEIGHT IS IN (CM)

```python
from sklearn.preprocessing import StandardScaler
```

```python
# apply gradient descent to only the input features so there is no need to standardise the y - aprt only the n
#only the need is to standardize the x - part

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
```

```python
# hame train dataset ka mean aur deviation yhn pr hi use krna h ! na ki test dataset ka to
# yhn fit_transform nahi use hoga. (PREVENTSV THE DATA LEAKAGE)
x_test=scaler.transform(x_test)
```

# APPLY LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()

# parameters like fit_intercept : it sees if u want intercept or not if set to false it won't add intercept
# if fit_intercept is true then it will normalize ) alr done)
#n_ jobs : kitne processors lane h use mein during training

#positive : to set all the coefficients positive 
#( if set to true)
```

```python
regression.fit(x_train,y_train)# agar yhn x_series use hota to error aa jata as independent wale ka 2-d hona jruri h
```

```python
regression.coef_ #y=a+bx hoti h equation naaa to ye b h
```

```python
regression.intercept_ # ye a hai
```

# plot the best fit line

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,regression.predict(x_train))
plt.show()
```

# predict for the test data

```python
y_pred=regression.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

```

# TO FIND THE RMSE , MAE , MSE

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

# TO FIND THE VALUES OF R^2

- COEFFICIENT OF DETERMINATION=SUM OF SQUARE OF RESIDUALS-TOTAL SUM OF SQUARES

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
print(score)
```
