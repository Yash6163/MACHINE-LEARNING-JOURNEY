## FEATURE ENGINEERING


```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df[['Date_of_Journey','Date','Month','Year']]
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df


```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df
```

```python
df['Arrvial_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrvial_min']=df['Arrival_Time'].str.split(':').str[1]
df
```

```python
df.info()
```

```python
df['Arrvial_hour']=df['Arrvial_hour'].astype(int)
df['Arrvial_min']=df['Arrvial_min'].astype(int)
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df
```

```python
df['Dept_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dept_min']=df['Dep_Time'].str.split(':').str[1]
df['Dept_hour']=df['Dept_hour'].astype(int)
df['Dept_min']=df['Dept_min'].astype(int)
df
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
import numpy as np
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,'4 stops':4})
df['Total_Stops']=df['Total_Stops'].fillna(1)
df
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df
```

```python
df['Duration_Hour']=df['Duration'].str.split(" ").str[0].str.split('h').str[0]
```

```python
df
```

```python
df['Duration_min']=df['Duration'].str.split(" ").str[1].str.split("m").str[0]
```

```python
df
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df['Duration_min']=df['Duration_min'].fillna(0)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encode=OneHotEncoder()
airline=encode.fit_transform(df[['Airline']]).toarray()
```

```python
enc_df=pd.DataFrame(airline,columns=encode.get_feature_names_out())
df=pd.concat([df,enc_df])

df
```
