```python
from sklearn.datasets import make_classification

x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.8],random_state=12)
```

##UNDERSTANDING SMOTE

- SKLEARN.DATASETS. (MAKE_CLASSIFICATION)
- IBMLEARN.OVERSAMPLING
- PANDAS

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df1
```

```python
df2=pd.DataFrame(y,columns=['target'])
df2
```

```python
final_data=pd.concat([df1,df2],axis=1)
final_data
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_data['f1'],final_data['f2'],c=final_data['target'])
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
x,y=oversample.fit_resample(final_data[['f1','f2']],final_data['target'])

x.shape
df1_new=pd.DataFrame(x,columns=['f1','f2'])
df2_new=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1_new,df2_new],axis=1)
final_df['target'].value_counts()
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
plt.show()
```
