```python
import numpy as np
```

```python
lst_marks=[10,20,100,60,60,70,8,6,77,22,88]
min,q1,median,q3,max=np.quantile(lst_marks,[0,0.25,0.5,0.75,1.0])
print(min,q1,median,q3,max)
```

```python
iqr=q3-q1
print(iqr)
```

```python
lower=q1-1.5*iqr
higher=q3+1.5*iqr
print(lower,higher)
```

```python
import seaborn as sns
sns.boxplot(lst_marks)
```
