```python
import pandas as pd
import numpy as np
np.random.seed(123)

n_samples=1000
class_0_ratio=0.8
n_class_0=int(n_samples*class_0_ratio)
n_class_1=n_samples-n_class_0

class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
    
})
class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1]*n_class_1
    
})

df=pd.concat([class_0,class_1]).reset_index(drop=True)
df
```

```python
df_min=df[df['target']==1]
df_max=df[df['target']==0]
```

```python
from sklearn.utils import resample

df_min_unsampled=resample(df_min,replace=True,n_samples=len(df_max),random_state=42)
df_min_unsampled
```

```python
##VERIFICATION OF CHANGES
df=pd.concat([df_max,df_min_unsampled])
df.value_counts('target')
```

```python
##DOWNCASTING : 

import pandas as pd
import numpy as np
np.random.seed(123)

n_samples=1000
class_0_ratio=0.8
n_class_0=int(n_samples*class_0_ratio)
n_class_1=n_samples-n_class_0

class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
    
})
class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1]*n_class_1
    
})

df=pd.concat([class_0,class_1]).reset_index(drop=True)
df

df_min=df[df['target']==1]
df_max=df[df['target']==0]
```

```python
from sklearn.utils import resample

df_max_unsampled=resample(df_min,replace=False,n_samples=len(df_min),random_state=42)
df_max_unsampled
df=pd.concat([df_max_unsampled,df_min])
df
df.shape
```
