```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# CREATING THE DATA

```python
x= 6 * np.random.rand(100,1)-3

# NP.RANDOM(100,1) MAKE ARRAY OF 100,1 SHAPE THEN THEN VALUES ARE NULTIPLIED BY 6 
# 3 IS SUBTRACTED .

y=0.5*x**2+1.5*x+2+np.random.randn(100,1)

# AB YHN P Y KI VALUES GENERATE KRI AND THEN F(X)+(VO GAUSIAN NORMAL RANDOM VALUE)
#( F(X)= 0.5X^2+1.5*X+2
```

```python
x,y
```

# CREATING THE SCATTER PLOT

```python
plt.scatter(x,y,color='green')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

# NOW CREATING THE TRAIN-TEST DATA SPLIT

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=42)
```

```python
x.shape,y.shape
```

# APPLY LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()

regression.fit(x_train,y_train)
```

# NOW CHECK THE PREFICTION CORRECTNESS

```python
from sklearn.metrics import r2_score
sscore=r2_score(y_test,regression.predict(x_test))
print(sscore)
```

# SEE THE ERROR VISUALIZATION


```python
plt.plot(x_train,regression.predict(x_train),color='blue')
plt.scatter(x_train,y_train,color='red')
plt.show()
```
