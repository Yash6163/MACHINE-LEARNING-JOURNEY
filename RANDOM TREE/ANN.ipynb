# ANN

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Sample input
X = np.array([[1, 0]])

# Build model
model = keras.Sequential([
    layers.Dense(2, activation='sigmoid', input_shape=(2,)),
    layers.Dense(1, activation='sigmoid')
])

# Initialize weights manually (optional for clarity)
model.set_weights([
    np.array([[0.5, -0.2],
              [0.3, 0.8]]),     # W1
    np.array([0.1, 0.1]),       # b1
    np.array([[0.7],
              [-1.2]]),         # W2
    np.array([0.1])             # b2
])

# Forward propagation (prediction only)
output = model.predict(X)

print("Forward Propagation Output:", output)
```

# BACK PROPAGATION

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers

# Data
X = tf.constant([[1.0, 0.0]])
y = tf.constant([[1.0]])

# Model
model = tf.keras.Sequential([
    layers.Dense(2, activation='sigmoid'),
    layers.Dense(1, activation='sigmoid')
])

# Loss function
loss_fn = tf.keras.losses.MeanSquaredError()

# Optimizer
optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)

# Training loop
for epoch in range(1000):

    with tf.GradientTape() as tape:
        predictions = model(X)
        loss = loss_fn(y, predictions)

    # Compute gradients (BACKPROP)
    gradients = tape.gradient(loss, model.trainable_variables)

    # Apply gradients
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

# Final output
print("Final Prediction:", model(X).numpy())
```


# FRONT USING FUNCTIONS

```python
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Model
model = Sequential([
    Dense(2, activation='relu', inputshape=(2,)),
    Dense(1, activation='sigmoid')
])

# Forward propagation (prediction)
X = np.array([[1, 0]])
print(model.predict(X))
```

# BACK USING FUNCTIONS

```python
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Data
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

# Model
model = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])

# Compile (defines backprop)
model.compile(optimizer='adam', loss='binary_crossentropy')

# Backpropagation happens here
model.fit(X, y, epochs=100, verbose=0)

# Output
print(model.predict(X))
```

# PCA

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Sample data
X = np.array([
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0]
])

# Step 1: Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

print("Reduced Data:\n", X_pca)
print("Explained Variance:", pca.explained_variance_ratio_)
```
