```python
# fill age nulls with median age
df['Age'].fillna(df['Age'].median(),inplace=True)

# mode for the type the contract
df['TypeofContact'].fillna(df['TypeofContact'].mode()[0],inplace=True)
#mode() does NOT return a single value.
#It returns a Series, because there can be more than one mode.


#median duration of pitch
df['DurationOfPitch'].fillna(df['DurationOfPitch'].median(),inplace=True)

# mode no of followers
df['NumberOfFollowups'].fillna(df['NumberOfFollowups'].mode()[0],inplace=True)

#mode for preferred property star
df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].mode()[0],inplace=True)

#median no of trips
df['NumberOfTrips'].fillna(df['NumberOfTrips'].median(),inplace=True)

#ode for the number of children visiting
df['NumberOfChildrenVisiting'].fillna(df['NumberOfChildrenVisiting'].mode()[0],inplace=True)

#median for monthly income
df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(),inplace=True)
```

#  PERFORMING THE TRAIN TEST SPLIT AND OHE AND BE

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['ProdTaken'],axis=1)
y=df['ProdTaken']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# now performing one hot encoding and binary encoding
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe=OneHotEncoder(drop='first')
# we use drop first kyuki we have to drop 1 column , as n collumn hai to n-1 column me hi data hoga encoded
sc=StandardScaler()
cat_feat = x.select_dtypes(include='object').columns.tolist()

# WE CAN ALSO DO THIS BY -
#cat_feat=[feat for feat in df.columns if df[feat].dtype=='O']

num_feat = x.select_dtypes(exclude='object').columns.tolist()

#visne ek object bna dia ab vo ek bari me appl krega
preprocessor=ColumnTransformer(
    [
        ("OneHotEncoder",ohe,cat_feat),
        ("StandardScaler",sc,num_feat)
    ]
)
```

```python
x_test
```

```python
x_train=preprocessor.fit_transform(x_train)
```

```python
pd.DataFrame(x_train)
```

```python
x_test=preprocessor.transform(x_test)

# to prevent dataleakage vrna fit_transform mer data will get to know ki test data bhi h 
```

# RANDOM FOREST MACHINE LEARNING ALGO TRAINING

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,r2_score,roc_auc_score
from sklearn.linear_model import RidgeCV
```

```python
models={
    "Decision Tree":DecisionTreeClassifier(),
    "Random Tree":RandomForestClassifier(),
    "Logistic Regression": LogisticRegression(),
    "NAIVE BAYES ": GaussianNB(),
    "SVC ": SVC(kernel="rbf"),
    "KNN" : KNeighborsClassifier(),
    "ADABoost":AdaBoostClassifier()
}
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    #now do test data 
    y_pred_train=model.predict(x_train)
    y_pred=model.predict(x_test)
    print(list(models.keys())[i])
    print("----------------------------\n")
    #test performence for train data
    print("accuracy_score : ",accuracy_score(y_train,y_pred_train))
    print("confusion matrix : ",confusion_matrix(y_train,y_pred_train))
    print("classification report",classification_report(y_train,y_pred_train))

    print("----------------------------\n\n\n")

    print(list(models.keys())[i])
    print("----------------------------\n")
    #test performence for test data
    print("accuracy_score : ",accuracy_score(y_test,y_pred))
    print("confusion matrix : ",confusion_matrix(y_test,y_pred))
    print("classification report",classification_report(y_test,y_pred))
    print("----------------------------\n")
```

# PCA

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Sample data
X = np.array([
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0]
])

# Step 1: Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

print("Reduced Data:\n", X_pca)
print("Explained Variance:", pca.explained_variance_ratio_)
```
