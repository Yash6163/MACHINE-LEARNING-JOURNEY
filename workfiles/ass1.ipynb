```python
def get_total(a,b,c):
    return a+b+c

def get_avg(a,b,c):
    return (a+b+c)/3

def calc_grade(mark):
    if mark>=90:
        return 'A+'
    elif mark>=85 and mark<=89 :
        return 'A'
    elif mark>=75 and mark<=84 :
        return 'B+'
    elif mark>=70 and mark<=74 :
        return 'B'
    elif mark>=65 and mark<=69 :
         return 'C+'
    elif mark >=60 and mark<=64 :
        return 'C'
    elif mark>=40 and mark<=59 :
        return 'D'
    else :
        return 'F'

student_data = []                       #[{'name' : yash , 'roll':18 , 'age': 18 , 'course':'btech'}]
n = int(input("ENTER NUMBER OF STUDENTS : "))
for i in range(0, n):
    dic = {}
    name = input(f"ENTER THE NAME OF STUDENT {i + 1} : ")
    roll = int(input(f"ENTER THE ROLL NO OF STUDENT {i + 1}: "))
    age = int(input(f"ENTER THE AGE OF THE STUDENT {i + 1} : "))
    course = input(f"ENTER THE COURSE NAME OF THE STUDENT {i + 1} : ")
    dic['name'] = name
    dic['roll'] = roll
    dic['age'] = age
    dic['course'] = course
    student_data.append(dic)

##input of each student's marks in 3 subjects (stored in a nested list)
sub_list = ("MATHS", "PHYSICS", "CS")
mark = []     #main list to store marks of n students                #[[10,20,20],[30,40,40],[60,70,80]]
grade=[]      # main list of grades of n students                    #[[A,A+,A],[A,B,C],[A,A+,B+]]

total_marks_of_each=[]      #total of each student stored              #[40,90,100]

avg_of_each=[]        #avg of each student stored                           #[10,10,10]

for i in range(0, n):
    stu = []
    grade_of_student=[]

    marks_ma = int(input(f"ENTER  THE MARKS IN MATHS OF STUDENT {i+1}: "))
    g_ma=calc_grade(marks_ma)
    grade_of_student.append(g_ma)

    marks_phy = int(input(f"ENTER THE MARKS IN PHYSICS OF STUDENT {i+1}: "))
    g_phy=calc_grade(marks_phy)
    grade_of_student.append(g_phy)

    marks_cs = int(input(f"ENTER THE MARKS IN CS OF STUDENT {i+1}: "))
    g_cs=calc_grade(marks_cs)
    grade_of_student.append(g_cs)

    stu.append(marks_ma)
    stu.append(marks_phy)
    stu.append(marks_cs)

    total=get_total(marks_ma,marks_phy,marks_cs)
    avg=get_avg(marks_ma,marks_phy,marks_cs)
    total_marks_of_each.append(total)  #appending the total of student
    avg_of_each.append(avg)            #appending the avg of the student

    mark.append(stu)                   #appending marks of student
    grade.append(grade_of_student)     #appending respective grade

final_dict={}
z=1
for person,mark,grade,total,avg in zip(student_data,mark,grade,total_marks_of_each,avg_of_each) :
    final_dict[f"student {z}"] = {
        "personal_info": person,
        "marks": mark,
        "grade": grade,
        "total": total,
        "avg": avg
    }
    z+=1

for student, details in final_dict.items():
    print(f"\n{student}:")
    print(f"  Personal Info: \n  NAME : {details['personal_info']['name']}\n  AGE:{details['personal_info']['age']}\n  ROLL NO :{details['personal_info']['roll']}\n  COURSE : {details['personal_info']['course']}")
    print("\n")
    print(f"  Marks:\n  {sub_list[0]}:{details['marks'][0]}\n  {sub_list[1]}:{details['marks'][1]}\n  {sub_list[2]}:{details['marks'][2]}")
    print("\n")
    print(f"  Grades:\n  {sub_list[0]}:{details['grade'][0]}\n  {sub_list[1]}:{details['grade'][1]}\n  {sub_list[2]}:{details['grade'][2]}")
    print("\n")
    print(f"  Total Marks: {details['total']}")
    print(f"  Average: {details['avg']:.2f}")
```
