```python
def square(n):
    for i in range(n):
        yield i**2
a=square(3)
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen=my_generator()
```

```python
for val in gen :
    print(val)
```

```python
def read_large_file(file_path):
    with open(file_path,'r')as file :
        for line in file :
            yield line 
file_path='largefile.txt'
for line in read_large_file(file_path):
    if(line[0] not in 'aeiou'):
        continue
    else:
        print(line.strip())
```
