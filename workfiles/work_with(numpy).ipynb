```python
import numpy as np
##create arrays using numpy
##create a 1d array
arr1=np.array([1,2,3,4,5,6])
arr1.reshape(2,3)
print(arr1.shape)
print(arr1)
arr2=np.array([1,2,3,4,5])
arr2.reshape(1,5)   ## 1 row and 5 columns
arr2=np.array([[1,2,3,4,5],[2,3,4,5,5]])
print(arr2.shape)
print(arr2)
```
