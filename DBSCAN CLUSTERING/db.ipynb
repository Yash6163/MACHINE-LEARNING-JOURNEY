```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# IMPORT A NEW FORM OF DATA
- BY USING MAKE_MOON WE GET DATA IN A CRESENT SHAPE FORMAT

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
x,y=make_moons(n_samples=250,noise=0.05)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

#  APPLY STANDARD SCALING AND TRAINING MODEL

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_scaled=ss.fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.3)
```

```python
dbscan.fit(x_scaled)
```

#  VERIFYING OUR PLOTS

```python
plt.scatter(x[:,0],x[:,1],c=dbscan.labels_)
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```
