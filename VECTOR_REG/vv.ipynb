```python
#idependent and dependent features
# if we do not apply the independent and dependent features there is a chance of data leakage
df.columns

X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y=df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
x_train
```

```python
y_train
```

```python
## Feature ENCODINF

#LABEL ENCODING 
# WHERE THER ARE 2 CATEOGRIES REPLACE BY 0 & 1

from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()

x_train['sex']=le1.fit_transform(x_train['sex'])
x_train['smoker']=le2.fit_transform(x_train['smoker'])
x_train['time']=le3.fit_transform(x_train['time'])

x_train
```

```python
le4=LabelEncoder()
le5=LabelEncoder()
le6=LabelEncoder()

x_test['sex']=le4.fit_transform(x_test['sex'])
x_test['smoker']=le5.fit_transform(x_test['smoker'])
x_test['time']=le6.fit_transform(x_test['time'])
x_test
```

```python
# now ab day pr one jhow encoding krna pdega kyuki 4 categories h
from sklearn.compose import ColumnTransformer  
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(drop='first'),[3])],remainder='passthrough')
# 3-> mtlb day column pr one hot encoding krna h
#drop first-> to avoid dummy variable trap mtlb first column jisse naye mile column ko drop krdo
#remaidner-> jo bache hue column h unko as it is rkhdo
ct.fit_transform(x_train)
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
x_train=ct.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=ct.transform(x_test)
# not using fit_transform bcoz we have to use the same encoding as train data
#mtlb jo encoding train data pr ki h usi ko test data pr apply krna h aur jisse ki avg mean ye sari values same rhe 
#vrna error k chances bdh jaynege
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='linear')
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print((classification_report(y_pred,y_pred)))
```
